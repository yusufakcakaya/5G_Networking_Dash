### Deman Driven Postalcode Data Results

```python
df = pd.read_csv('data/demand_driven_postcode_data_results.csv')
```

```python
limit = df.query("spectrum_limitations == 'all_spectrum'")
```

```python
geo_mar_bar = px.histogram(limit, x="geotype", y="Capacity.surplus", title="Geotype Margin Bar Plot",
labels={'pop':'strategy'}, height=400)
geo_mar_bar.update_layout(title_x=0.5)

geo_mar_bar.show()
```

```python
geo_mar_bar = px.histogram(limit, x="geotype", y="Existing.capacity", title="Geotype Margin Bar Plot",
labels={'pop':'strategy'}, height=400)
geo_mar_bar.update_layout(title_x=0.5)

geo_mar_bar.show()
```
